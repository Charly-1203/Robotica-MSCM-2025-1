## Objetivo

El objetivo de esta práctica es que el alumno comprenda, interprete y modifique la información básica de los formatos de Universal Robot Description File (URDF).

## Introducción

En el marco de la robotica actual Se utilizan varios métodos para diseñar y aplicar el manipulador.Entre ellos, el más eficiente en términos de coste y tiempo es una simulación [[1]](#1) en consecuencia, resulta fundamental tener una representacion precisa del sistema para poder simularlo, controlarlo y visualizarlo. El Formato URDF (Universal Robot Description File) por sus siglas en ingles, se presenta como una herramienta escencial en el entonro ROS para describir la estructura y propiedades fisicas de un robot, a través de este archivo, se puede definir la geometría, dinámica y cinemática de cada uno de sus componentes, permitiendo una interacción más intuitiva y controlada con el entorno de simulación, en este caso, Rviz. Cabe recalcar que es imperativo entender la atingencia de las diferentes partes que conforman la descripcion de un robot y los entornos que se utilizan para simularlo, por lo que a lo largo de esta practica se abordaran los conceptos basicos que propiciaran dicho cometido de manera exitosa.

### Previo

#### Definición de URDF (Unified Robot Description Format):

URDF es un formato basado en XML utilizado para describir la estructura física de los robots en ROS. Este formato especifica los enlaces (links) y las juntas (joints) que forman un robot, definiendo geometría, masa, inercia y las conexiones entre las partes móviles. Es fundamental para la visualización y simulación de robots en herramientas como RViz y Gazebo.
#### Tipos de juntas en URDF:
 
* Revolute: articulación con un grado de libertad de revolución en torno a 
un eje con un rango limitado de giro, que vienen delimitados por un rango 
superior e inferior. 
* Fixed: articulación fija en la que no existe ningún grado de libertad. Este tipo de articulación no requiere especificar las etiquetas:
```xml

 * <axis>
 * <calibration>
 * <dynamics>
 * <limits>
 * ni safety_controller.
    
```
* Continuos: una articulación similar a la articulación revolute pero sin límites de rango de giro. 
* Prismatic: una articulación de un grado de libertad de desplazamiento linear sobre un eje dentro de un rango limitado. 
* Planar: articulación con 2 grados de libertad que permite el movimiento en un plano perpendicular al eje especificado. 
* Floating: movimiento libre en los 6 grados de libertad

#### Elementos primitivos en ROS (para definir robots): 
Los elementos "primitivos" utilizados en URDF para describir robots incluyen:

* Links: Son las partes rígidas del robot. Cada link puede tener propiedades como masa, inercia, y representación visual. Se definen dentro de la estructura URDF y son cruciales para la modelación del robot

* Joints: Conectan los enlaces y permiten ciertos tipos de movimiento entre ellos. Existen varios tipos de juntas, como revolute (rotación), prismatic (deslizamiento), fixed (sin movimiento), floating (libre en todos los grados de libertad), y continuous (rotación continua). Cada tipo de joint tiene atributos que definen su comportamiento, como límites de movimiento y ejes de rotación o deslizamiento

* Geometry: Se refiere a las formas que representan visualmente los enlaces del robot. Esto puede incluir primitivas como cajas, cilindros, esferas, y mallas 3D. Estas definiciones son esenciales para la visualización y simulación del robot en entornos como RViz o Gazebo

* Visual: Definen cómo se verá el robot, es decir la apariencia (Color, tamaño etc). 

* Collision: Esta propiedad y cómo interactuará con otros objetos en el entorno, es decir define cómo se manejan las interacciones físicas con otros objetos

* Inertial: Define las propiedades inerciales de un link, que son fundamentales para simular correctamente el comportamiento dinámico del robot en acción


#### Definición de nodo y tópico en ROS 2:
  
* Nodo: Es una unidad de procesamiento básica en ROS que ejecuta una parte del sistema de control. Los nodos pueden publicar y suscribirse a mensajes, realizar cálculos y actuar como interfaces con sensores y actuadores.
      
* Tópico: Es un canal de comunicación asíncrono entre nodos. Los mensajes se publican en un tópico, y otros nodos pueden suscribirse a estos tópicos para recibir los datos correspondientes
      
#### Definición de RViz:

* RViz es una herramienta de visualización 3D en ROS que permite mostrar datos de sensores, modelos de robots y la información del entorno en tiempo real. Es muy útil para depurar y monitorear el estado y movimientos de un robot, ofreciendo diversas opciones de visualización como nubes de puntos, trayectorias, modelos de robots y más.


## Desarrollo

Como primer parte de la practica se realizo la creacion de un eslabon base que se declaro fija como tierra y como segunda instancia se declaro un link_1 unido a la base.

1. Declaración inicial del archivo XML: Especifica la versión de XML y la codificación del archivo. Aquí se define que el documento utiliza la versión 1.0 de XML y está codificado en UTF-8.

```xml

<?xml version="1.0" encoding="utf-8"?>

```
2. Definición del robot:Inicia la definición de un robot con el nombre basic_link.

```xml
<robot name="basic_link">
```
3. Definición del mundo (link base):Define un "link" o eslabón que representa el mundo o el entorno donde el robot está situado. No tiene propiedades, solo sirve de referencia.

```xml
<link name="world">
</link>

```
4. Definición de la junta fija entre el mundo y la base:Crea una junta fija que conecta el mundo (world) con la base del robot (base_link), indicando la posición y orientación con coordenadas (0,0,0) y rotaciones (0,0,0).

```xml
<joint name="world_to_base_link" type="fixed">
    <parent link="world"/>
    <child link="base_link"/>
    <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
</joint>
```

5. Visualizacion del eslabón base: Define las propiedades visuales de la base, como la geometría (un cubo de tamaño 0.1x0.12x0.15), posición y color (gris)
   
```xml
<visual>
    <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
    <geometry>
        <box size="0.1 0.12 0.15"/>
    </geometry>
    <material name="grey">
        <color rgba="0.5 0.5 0.5 1.0"/>
    </material>
</visual>

```
6. Definición de la colisión del eslabón base: Similar a la sección visual, define cómo el eslabón interactúa con otros objetos en términos de colisión.
```xml

<collision>
    <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
    <geometry>
        <box size="0.1 0.12 0.15"/>
    </geometry>
</collision>

```
7. Definición de las propiedades inerciales:Descripción: Define la masa (0.5) y las propiedades de inercia, lo que es importante para simular correctamente la dinámica del eslabón.
```xml

<collision>
    <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
    <geometry>
        <box size="0.1 0.12 0.15"/>
    </geometry>
</collision>

```
Con lo anterior tenemos la base fijada a la tierra

![base](imagenes_p1/base.png) 

Despues se procede a la creacion del primer link de manera homologa a la metodologia seguida para la programacion del base_link y de su junta 

8. Creacion del Link_1 y de su junta 
 ```xml

   <link name="link_1">
        <!--Descripcion de la parte visual del eslabon-->
        <visual>
            <!--Posicion y orientación de origen del eslabon-->
            <origin xyz="0.025 0.0 0.25" rpy="0.0 0.0 0.0"/>
            <!--Definición de la geometria del eslabón-->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>
            <!--Definicion del color del eslabon-->
            <material name="red">
                <color rgba="0.5 0.0 0.0 1.0" />
            </material>      
        </visual>
        <!--Descripcion de las relaciones de colision-->
        <collision>
            <!--Posicion y orientacion de origen del eslabon-->
            <origin xyz="0.025 0.0 0.25" rpy="0.0 0.0 0.0"/>
            <!--Definicion de la geometria del eslabon-->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>          
        </collision>
        <!--Descripcion de la inercial del eslabon-->
        <inertial>
            <!--masa del eslabon-->
            <mass value="0.3"/>
            <!--Ubicacion del sistema inercial asociado al centro de masa-->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!--Matriz de inercia en el centro de masa-->
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>

   <joint name="base_link_to_link_1" type="revolute">
    <parent link="base_link"/>
    <child link="link_1"/>
    <axis xyz="1 0 0"/>
    <limit lower="-3.14159" upper="3.14159" velocity="50.0" effort="1000.0"/>
    <origin xyz="0.05 0 0.15" rpy="0 0 0"/>
   </joint>
   
   ```

Para la segunda parte de la practica basicamente se sigue la misma metodologia para la creacion del segundo eslabon 

10. Creacion del Link_2 y de su junta 
```xml
 
 <!--Definicion del Segundo Link-->
    <link name="link_2">
        <!--Descripcion de la parte visual del eslabon-->
        <visual>
            <!--Posicion y orientación de origen del eslabon-->
            <origin xyz="0.025 0.00 0.25" rpy="0.0 0.0 0.0"/>
            <!--Definición de la geometria del eslabón-->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>
            <!--Definicion del color del eslabon-->
            <material name="purple">
                <color rgba="0.187 0.143 0.206 1.0" />
            </material>      
        </visual>
        <!--Descripcion de las relaciones de colision-->
        <collision>
            <!--Posicion y orientacion de origen del eslabon-->
            <origin xyz="0.025 0.05 0.25" rpy="0.0 0.0 0.0"/>
            <!--Definicion de la geometria del eslabon-->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>          
        </collision>
        <!--Descripcion de la inercial del eslabon-->
        <inertial>
            <!--masa del eslabon-->
            <mass value="0.3"/>
            <!--Ubicacion del sistema inercial asociado al centro de masa-->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!--Matriz de inercia en el centro de masa-->
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>


   <joint name="link_1_to_link_2" type="revolute">
    <parent link="link_1"/>
    <child link="link_2"/>
    <axis xyz="1 0 0"/>
    <limit lower="-3.14159" upper="3.14159" velocity="50.0" effort="1000.0"/>
    <origin xyz="0.05 0 0.5" rpy="0 0 0"/>
   </joint>
   
  ```
11. Definición de propiedades en Gazebo:Especifica los materiales para simular en el entorno de Gazebo, donde el base_link será gris y el link_1 rojo.
 
```xml

<gazebo reference="base_link">
    <material>Gazebo/Grey</material>
</gazebo>

<gazebo reference="link_1">
    <material>Gazebo/Red</material>
</gazebo>

```
12. Definición de la transmisión y control del robot:Define cómo se controlará el robot, con una transmisión simple que conecta la junta base_link_to_link_1 con un motor actuador.

```xml
<transmission name="trans_1">
    <type>transmission_interface/SimpleTransmission</type>
    <joint name="base_link_to_link_1">
        <hardwareInterface>hardware_interface/PositionJointInterface</hardwareInterface>
    </joint>
    <actuator name="motor_1">
        <hardwareInterface>hardware_interface/PositionJointInterface</hardwareInterface>
        <mechanicalReduction>1</mechanicalReduction>
    </actuator>
</transmission>

```
13. Plugins de Gazebo para la simulación:Añade dos plugins a Gazebo, uno para publicar los estados de las juntas (joint_state_publisher_controller) y otro para controlar la trayectoria de las poses de las juntas (joint_pose_trajectory_controller).

```xml

<gazebo>
    <plugin name="joint_state_publisher_controller" filename="libgazebo_ros_joint_state_publisher.so">
        <update_rate>10</update_rate>
        <joint_name>base_link_to_link_1</joint_name>
    </plugin>
</gazebo>

<gazebo>
    <plugin name="joint_pose_trajectory_controller" filename="libgazebo_ros_joint_pose_trajectory.so">
        <update_rate>2</update_rate>
    </plugin>
</gazebo>

```
14. Cierre del robot: Finalmente se cierra la estructura del robot
```xml

</robot>

```
